```python
import random, pygame, sys
import math
from pygame.locals import *

pygame.init()
L, H = 1000, 680
CX, CY = int(L/2), int(H/2)
DISPLAYSURF = pygame.display.set_mode((L, H))
FPSCLOCK = pygame.time.Clock()

position = {"x":-CX, "y":-CY}
mouse = (CX, CY + 100)
dx = "pas bouger"
dy = "pas bouger"

monde = (2000, 2000)

squares = []
pentagones = []

boulets = []

for i in range(0,100):
    squares.append((random.random()*monde[0]-monde[0]/2, random.random()*monde[1]-monde[1]/2))

def rotate(center, point, angle):
    cx, cy = center[0], center[1]
    x, y = point[0], point[1]
    cosx, sinx = math.cos(angle), math.sin(angle)
    x, y = x - cx, y - cy
    return (
        x * cosx - y * sinx + cx,
        x * sinx + y * cosx + cy
    )

def rotate_rect(center, points, angle):
    res = []
    for p in points:
        res.append(rotate(center, p, angle))
    return res

def angle(a, b):
    x1, y1 = a[0], a[1]
    x2, y2 = b[0], b[1]
    return math.atan2(y2-y1, x2-x1)
    
def afficheMoi():
    x, y = CX, CY
    a = angle((x,y), mouse) - math.pi / 2
    l = 3
    rect = [(x-5, y), (x+5, y), (x+5, y+40), (x-5, y+40)]
    rect2 = [(x-5-l, y), (x+5+l, y), (x+5+l, y+40+l), (x-7, y+40+l)]
    canon = rotate_rect((x,y), rect, a)
    canon2 = rotate_rect((x,y), rect2, a)
    pygame.draw.polygon(DISPLAYSURF, (60,60,180), canon2)
    pygame.draw.polygon(DISPLAYSURF, (180,183,187), canon)
    pygame.draw.circle(DISPLAYSURF, (60,60,180), (x,y), 22)
    pygame.draw.circle(DISPLAYSURF, (124,176,253), (x,y), 18)

def afficheMonde():
    l, h = monde[0], monde[1]
    x = position["x"]
    y = position["y"]
    pygame.draw.rect(DISPLAYSURF, (150,200,100), pygame.Rect(int(-l/2)-x-500, -int(h/2)-y-500, l+1000, h+1000))
    pygame.draw.rect(DISPLAYSURF, (180,220,120), pygame.Rect(int(-l/2)-x, -int(h/2)-y, l, h))
    for i in range(int(-l/2), int(l/2)+1, 50):
        pygame.draw.line(DISPLAYSURF, (255,255,255), (i-x, int(-h/2)-y), (i-x, int(h/2)-y))
    for j in range(int(-h/2), int(h/2)+1, 50):
        pygame.draw.line(DISPLAYSURF, (255,255,255), (int(-l/2)-x, j-y), (int(l/2)-x, j-y))

def afficheObstacle():
    x = position["x"]
    y = position["y"]
    for i in squares:
        pygame.draw.rect(DISPLAYSURF, (210,50,50), pygame.Rect(i[0]-x, i[1]-y, 20, 20))

def afficheBoulets():
    for b in boulets:
        pygame.draw.circle(DISPLAYSURF, (50,50,253), (int(b[0]),int(b[1])), 7)

def collisions():
    x = position["x"]
    y = position["y"]
    for i in squares:
        for j in boulets:
            if pygame.Rect(i[0]-x, i[1]-y, 20, 20).colliderect(pygame.Rect(j[0]-7, j[1]-7, 14, 14)):
                squares.remove(i)
        
def terminate():
    pygame.quit()
    sys.exit()
    
while True:
    
    canon = "nepastirer"
    
    # Evenements
    for event in pygame.event.get(): # event handling loop
        if event.type == QUIT:
            terminate()
        elif event.type == KEYDOWN:
            if (event.key == K_LEFT or event.key == K_a) :
                dx = "gauche"
            elif (event.key == K_RIGHT or event.key == K_d):
                dx = "droite"
            elif (event.key == K_UP or event.key == K_w):
                dy = "haut"
            elif (event.key == K_DOWN or event.key == K_s):
                dy = "bas"
            elif (event.key == K_SPACE):
                canon = "tirer"
        elif event.type == KEYUP:
            if (event.key not in (K_LEFT, K_RIGHT, K_UP, K_DOWN)):
                continue
            elif (event.key == K_LEFT or dx == "gauche") :
                dx = "pas bouger"
            elif (event.key == K_RIGHT or dx == "droite"):
                dx = "pas bouger"
            elif (event.key == K_UP or dy == "haut"):
                dy = "pas bouger"
            elif (event.key == K_DOWN or dy == "bas"):
                dy = "pas bouger"
        elif event.type == MOUSEBUTTONDOWN:
            canon = "tirer"

    mouse = pygame.mouse.get_pos()
    
    # Calculs
    if canon == "tirer":
        boulets.append([CX,CY, angle((CX,CY), mouse)])
    
    # Nouvelles position
    vitesse = 2
    if dy == "haut":
        position["y"] = position["y"] - vitesse
    if dy == "bas":
        position["y"] = position["y"] + vitesse
    if dx == "gauche":
        position["x"] = position["x"] - vitesse
    if dx == "droite":
        position["x"] = position["x"] + vitesse
    newb = []
    for b in boulets:
        b[0] = b[0] + 5 * math.cos(b[2])
        b[1] = b[1] + 5 * math.sin(b[2])
        if b[0] < L and b[0] > 0 and b[1] < H and b[1] > 0:
            newb.append(b)
    boulets = newb
    
    # Collisions ?
    collisions()
    
    # Dessiner
    
    DISPLAYSURF.fill((255,255,4))
    
    afficheMonde()
    afficheObstacle()
    afficheBoulets()
    afficheMoi()
    
    pygame.display.update()
    
    FPSCLOCK.tick(100)
    
```
